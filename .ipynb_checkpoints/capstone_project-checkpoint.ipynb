```python
# cols_detect_outliers = ['LotFrontage', 'LotArea', 'MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'GrLivArea']
def detect_outliers(df, threshold=3):
    return df[(np.abs(stats.zscore(df)) < 3).all(axis=1)].index

def preprocess(data_frame, is_train=True):
    df = data_frame.copy()
    df = df.drop(['Id'], axis=1, inplace=True)
    # Fill None values for dtype object cells    
    objects = [feature for feature in df.columns if df[feature].dtype==object]
    df.update(df[objects].fillna('None'))

    df['LotFrontage'] = df.groupby('Neighborhood')['LotFrontage'].transform(lambda x: x.fillna(x.median()))

    numeric_dtypes = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
    numerics = []
    for i in df.columns:
        if df[i].dtype in numeric_dtypes:
            numerics.append(i)
    df.update(df[numerics].fillna(0))
    features = df.drop(['Utilities', 'Street', 'PoolQC',], axis=1)

    df['YrBltAndRemod']=df['YearBuilt']+features['YearRemodAdd']
    df['TotalSF']=df['TotalBsmtSF'] + df['1stFlrSF'] + features['2ndFlrSF']

    df['Total_sqr_footage'] = (df['BsmtFinSF1'] + df['BsmtFinSF2'] +
                                    features['1stFlrSF'] + df['2ndFlrSF'])

    df['Total_Bathrooms'] = (df['FullBath'] + (0.5 * features['HalfBath']) +
                                  df['BsmtFullBath'] + (0.5 * df['BsmtHalfBath']))

    features['Total_porch_sf'] = (df['OpenPorchSF'] + df['3SsnPorch'] +
                                  df['EnclosedPorch'] + features['ScreenPorch'] +
                                  df['WoodDeckSF'])
    
    df['haspool'] = df['PoolArea'].apply(lambda x: 1 if x > 0 else 0)
    features['has2ndfloor'] = df['2ndFlrSF'].apply(lambda x: 1 if x > 0 else 0)
    df['hasgarage'] = df['GarageArea'].apply(lambda x: 1 if x > 0 else 0)
    df['hasbsmt'] = df['TotalBsmtSF'].apply(lambda x: 1 if x > 0 else 0)
    df['hasfireplace'] = df['Fireplaces'].apply(lambda x: 1 if x > 0 else 0)



    if is_train:
        train = train[train.GrLivArea < 4500]
        train.reset_index(drop=True, inplace=True)
        train["SalePrice"] = np.log1p(train["SalePrice"])
        y = train['SalePrice'].reset_index(drop=True)


        skew_features = df[numerics].apply(lambda x: skew(x)).sort_values(ascending=False)

        high_skew = skew_df[skew_df > 0.5]
        skew_index = high_skew.index

        for i in skew_index:
            df[i] = boxcox1p(df[i], boxcox_normmax(df[i] + 1))

        final_df = pd.get_dummies(features).reset_index(drop=True)

        outliers = [30, 88, 462, 631, 1322]
        X = X.drop(X.index[outliers])
        y = y.drop(y.index[outliers])
        
        overfit = []
        for i in X.columns:
            counts = X[i].value_counts()
            zeros = counts.iloc[0]
            if zeros / len(X) * 100 > 99.94:
                overfit.append(i)

        overfit = list(overfit)
        X = X.drop(overfit, axis=1)
        X_sub = X_sub.drop(overfit, axis=1)
```

# Model Evaluation and Validation

```python
kfolds = KFold(n_splits=10, shuffle=True, random_state=42)

def root_mean_squared_error(y, y_pred):
    return np.sqrt(mean_squared_error(y, y_pred))

def cv_root_mean_squared_error(model, X=X, y=y):
    rmse = np.sqrt(-cross_val_score(model, X, y, scoring="neg_mean_squared_error", cv=kfolds))
    return (rmse)
```

```python
alphas_alt = [14.5, 14.6, 14.7, 14.8, 14.9, 15, 15.1, 15.2, 15.3, 15.4, 15.5]
alphas2 = [5e-05, 0.0001, 0.0002, 0.0003, 0.0004, 0.0005, 0.0006, 0.0007, 0.0008]
e_alphas = [0.0001, 0.0002, 0.0003, 0.0004, 0.0005, 0.0006, 0.0007]
e_l1ratio = [0.8, 0.85, 0.9, 0.95, 0.99, 1]
```

```python
ridge = make_pipeline(RobustScaler(), RidgeCV(alphas=alphas_alt, cv=kfolds))
lasso = make_pipeline(RobustScaler(), LassoCV(max_iter=1e7, alphas=alphas2, random_state=42, cv=kfolds))
elasticnet = make_pipeline(RobustScaler(), ElasticNetCV(max_iter=1e7, alphas=e_alphas, cv=kfolds, l1_ratio=e_l1ratio))                                
svr = make_pipeline(RobustScaler(), SVR(C= 20, epsilon= 0.008, gamma=0.0003,))
```

```python
gbr = GradientBoostingRegressor(n_estimators=3000, learning_rate=0.05, max_depth=4, 
                                max_features='sqrt', min_samples_leaf=15, 
                                min_samples_split=10, loss='huber', random_state =42)   
```

```python
lightgbm = LGBMRegressor(objective='regression', 
                                       num_leaves=4,
                                       learning_rate=0.01, 
                                       n_estimators=5000,
                                       max_bin=200, 
                                       bagging_fraction=0.75,
                                       bagging_freq=5, 
                                       bagging_seed=7,
                                       feature_fraction=0.2,
                                       feature_fraction_seed=7,
                                       verbose=-1,
                                       )
```

```python
xgboost = XGBRegressor(learning_rate=0.01,n_estimators=3460,
                                     max_depth=3, min_child_weight=0,
                                     gamma=0, subsample=0.7,
                                     colsample_bytree=0.7,
                                     objective='reg:linear', nthread=-1,
                                     scale_pos_weight=1, seed=27,
                                     reg_alpha=0.00006)
```

```python
stack_gen = StackingCVRegressor(regressors=(ridge, lasso, elasticnet, gbr, xgboost, lightgbm),
                                meta_regressor=xgboost,
                                use_features_in_secondary=True)
```

```python
score = cv_root_mean_squared_error(ridge)
score = cv_root_mean_squared_error(lasso)
print("LASSO: {:.4f} ({:.4f})\n".format(score.mean(), score.std()), datetime.now(), )

score = cv_root_mean_squared_error(elasticnet)
print("elastic net: {:.4f} ({:.4f})\n".format(score.mean(), score.std()), datetime.now(), )

score = cv_root_mean_squared_error(svr)
print("SVR: {:.4f} ({:.4f})\n".format(score.mean(), score.std()), datetime.now(), )

score = cv_root_mean_squared_error(lightgbm)
print("lightgbm: {:.4f} ({:.4f})\n".format(score.mean(), score.std()), datetime.now(), )

score = cv_root_mean_squared_error(gbr)
print("gbr: {:.4f} ({:.4f})\n".format(score.mean(), score.std()), datetime.now(), )

score = cv_root_mean_squared_error(xgboost)
print("xgboost: {:.4f} ({:.4f})\n".format(score.mean(), score.std()), datetime.now(), )
```
